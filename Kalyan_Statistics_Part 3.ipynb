## How to choose which type of graph to use

**Line-Graph:** Points connected by lines to show trends. Line graphs are used to track changes over short or long periods of time. When smaller changes exists, line graphs are better than bar graphs. Line graphs are used to compare changes over same period of time for more than one one group. Ex. Hear rate in 24 hours.

**Pie-Chart:** They are used when comparing parts of whole. Whole pie represents 100%, size of each part is proportional to its percentage. They do not show changes over time. (Ice cream sales in each month in a year)

**Bar-graph:** Bar charts are used to compare things between different groups or track changes over time. They are best when changes are larger. Ex. Military budget for past 10 years (shows each bar for each year)

**Scatter Plot (XY) plot:** Shows relationship between two sets of data. Ex. Temparature vs Ice cream sales

### What happens to mean when you remove outlier.

**Removing outlier from Lowest Data point:** Mean goes up. Median may/may not change.

**Removing outlier from Highest Data point:** Mean comes down. Median may/may not change.

Outlier not only affects mean. It also affects Standard Deviation.

**Standard Deviation:** It is measure of spread. It is Square root of variance. **Variance** is average squared deviation from mean. SD measures exactly how far points lie away from mean. 


 High SD means data is widely spread. Lower SD means data is narrowly spread.  

 Standard deviation is only used to measure spread or dispersion around the mean of a data set.

 Standard deviation is never negative.

 If all values of a data set are the same, the standard deviation is zero (because each value is equal to the mean).
