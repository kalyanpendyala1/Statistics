# Confidence Intervals

When estimating a population parameter using a sample statistic it never going to be perfect, there will be always some error. We express that error or uncertainity using an *Interval Estimate.*

**Point Estimate +- Margin of Error**


Standard Error of Mean is the standard deviation of Sampling distribution. To find SEM we need to know two things *1.Population Standard Deviation, 2.Sample Size*



Most often we do not know the population standard deviation we have to estimate it. For any population standard deviation increasing the sample size (No. of observations in a sample) reduces the standard error or SD of sampling distribution. 

### Point Estimate vs Interval Estimate

Point estimate is a single number that estimates the exact value of the population parameter of interest. 

Interval estimate includes range of possible values that are likely to include the actual population parameter.



When interval estimate is associated with a degree of confidence that it includes the population parameter, it is referred as **Confidence Interval.**

95% confidence interval is mostly used. In 95% confidence interval, we left with 5% that is evenly divided between both the tails. The area in the tails is denoted with alfa. it is alfa/2 on each side.

If we know the population standard deviation, we treat the sampling distribution curve as normal distribution curve. By doing so we can assign Z-Scores to upper and lower boundary of the 95% interval. After assigning the Z-scores calculate (positive-negative). You will get the probability for 95%.



Standard Error is SD of sampling distribution. SEM is dependent on the Sample size. If the sample size is small we get a smaller and wider curve. If the sample size is large, we will get taller curve and most of the observations are closer to the mean.

Samples of the same size have same standard error.



above picture show the 95% confidence interval curve, it shows that 95% of the sample means are inside that interval. Sometimes sample means may be outside the interval, like 5th mean, they count to other 5% which are outside (in the tails). thats why we have 95% confidence interval. 

### Interpretation

Randomness lies in the sample means not the population mean. It is the probability of obtaining the representative sample. 

The proportion of samples, size n, for which our estimate, the sample mean x_bar, is within  a certain distance +- of the true population mean.

The sample mean is either within +-interval of the true population mean, or it is not. 

#### Very common Error

The confidence interval is not the probability that the population mean lies within the interval.

"There is 95% probability the population mean is within the interval of _____" **wrong**

"There is 95% probability that the sample mean is within ______ so and so distance of population mean" (**Right**)

There is randomness to sample means not the population mean.

## Confidence Interval when sigma (Population SD) is unknown

Here we dont have the population standard deviation, we have to estimate it. For any population SD or its estimate, increasing the sample size reduces the standard error.

Confidence interval will be affected by the Standard Deviation, sample size and level of confidence we are satisfied with. (check formula, we have sample size, SD and level of confidence)



When estimating a mean score or a proportion from a single sample, the number of independent observations is equal to the sample size minus one. Hence, the distribution of the t statistic from samples of size 8 would be described by a t distribution having 8 - 1 or 7 degrees of freedom. Similarly, a t distribution having 15 degrees of freedom would be used with a sample of size 16.

The t distribution approaches the normal distribution as the degrees of freedom increase.
