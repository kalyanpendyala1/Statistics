# Hypothesis Testing

Hypothesis Testing is process of reaching out to conclusions based on the sample information. In particular, evaluating the hypothesis based on claims.

It is Inferential Statistics method because information from sample is used to make inferences about population.

### Null hypothesis

First step of Hypothesis testing is forming a "Null Hypothesis" (H sub Naught). The Null hypothesis is statement about the value of population parameter and is put up for testing.  

The Null hypothesis can be accepted or rejected. 

Null hypothesis is assumed to be true unless we have statistically overwhelming evidence to the contrary (Opposite).

### Alternative Hypothesis

Denoted by H_sub_1.

Alternative hypothesis is a statement that holds "*if the null hypothesis is False*".

For a given test Null and Alternative hypothesis include all the possible values of the population parameter, so either must be false.

Null and Alternative hypothesis are opposite and they are mutually exclusive.

### Directional Claim

Directional Claim or statement holds that a population parameter is greater than (>), at least (>=), no more than (<=), less than (<) some quantity.

**Example:** The president of a company that manufactures central home air conditioning units has told an investigative reporter that at least 85% of its
homeowner customers claim to be “completely satisfied” with the overall purchase experience.

### Non-Directional Claim

Non-Directional Claim or statement says that parameter equal to some quantity.

**Example:** Titus Walsh, the director of a municipal transit authority, claims that 35% of the system’s ridership consists of senior citizens.

### One tail Test

Directional claims results in one-tail test, where null hypothesis are rejected by extreme results in one direction only.

### Two tail Test

Non-directional claims involves Two tailed tests, in which null hypothesis can be rejected by extreme result occuring in both directions.

In the above figure there are two regions, One is white region that is called region of acceptence and region that is shaded in black colour is called region of rejection  or **Critical Region**.

### Differences between One-Tailed and Two-Tailed tests

In one tail test, Alternative hypothesis is Directional. Whereas in two tail test Alternative hypothesis is Non-directional. 

In one tail test Critical region(region of rejection) lies only in one tail, Where in two tail test critical region (region of regection) lies in both the tails.

One tailed test determines if there is a relationship between variables in single direction. Two-tailed test determines if there is a relationship between variables in either direction.

When an alternative hypothesis has ‘≠’ sign, then a two-tailed test is performed. In contrast, when an alternative hypothesis has ‘> or <‘ sign, then one-tailed test is carried out.

**Alpha levels** sometimes called "Significance levels" It is probability of making the wrong decision when null hypothesis is true. One tail test has entire alpha level in one tail (either left or right). A two tailed test splits your alpha level in half.

### Assumptions vs Claim

When formulating Statistical Hypothesis ask yourself that you are testing an "assumption or Status Quo", that is already exist? (water bottle). Or you are testing a claim or assertion beyond what we know (Hybrid car).

### Errors in Hypothesis Testing

Whenever we reject Null hypothesis, there is a chance that we have made a mistake - i.e., that is we have rejected a true statement.

Rejecting a true null hypothesis is referred as **"Type I Error"**.

The probability of making such error is referred as **"Significance level"**.
It is denoted by **alpha (α)**

Failing to reject a false null hypotheis "**Type II Error**".

Probability of making Type II Error is denoted by **beta (β)**.

If we accepted true null hypothesis and rejected false null hypothesis, we acted correctly.

The probability of rejecting a false null hypothesis is "**Power of the Test**"

#### Trade-off

In Hypothesis testing there is tradeoff between Type I and Type II errors. For a given sample size, reducing the probability of Type I Error increases probability of Type II Error and vice-versa.
