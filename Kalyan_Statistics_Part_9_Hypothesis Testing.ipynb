# Hypothesis Testing

Hypothesis Testing is process of reaching out to conclusions based on the sample information. In particular, evaluating the hypothesis based on claims.

It is Inferential Statistics method because information from sample is used to make inferences about population.

### Null hypothesis

First step of Hypothesis testing is forming a "Null Hypothesis" (H sub Naught). The Null hypothesis is statement about the value of population parameter and is put up for testing.  

The Null hypothesis can be accepted or rejected. 

Null hypothesis is assumed to be true until we prove alternate hypothesis to be true.

### Alternative Hypothesis

Denoted by H_sub_1.

Alternative hypothesis is a statement that holds "*if the null hypothesis is False*".

For a given test Null and Alternative hypothesis include all the possible values of the population parameter, so either must be false.

Null and Alternative hypothesis are opposite and mutually exclusive.

### Directional Claim

Directional Claim or statement holds that a population parameter is greater than (>), at least (>=), no more than (<=), less than (<) some quantity.

**Example:** The president of a company that manufactures central home air conditioning units has told an investigative reporter that at least 85% of its
homeowner customers claim to be “completely satisfied” with the overall purchase experience.

### Non-Directional Claim

Non-Directional Claim or statement says that parameter equal to some quantity.

**Example:** Titus Walsh, the director of a municipal transit authority, claims that 35% of the system’s ridership consists of senior citizens.

### One tail Test

Directional claims results in one-tail test, where null hypothesis are rejected by extreme results in one direction only.

### Two tail Test

Non-directional claims involves Two tailed tests, in which null hypothesis can be rejected by extreme result occuring in both directions.

In the above figure there are two regions, One is white region that is called region of acceptence and region that is shaded in black colour is called region of rejection  or **Critical Region**.

### Differences between One-Tailed and Two-Tailed tests

In one tail test, Alternative hypothesis is Directional. Whereas in two tail test Alternative hypothesis is Non-directional. 

In one tail test Critical region(region of rejection) lies only in one tail, Where in two tail test critical region (region of regection) lies in both the tails.

One tailed test determines if there is a relationship between variables in single direction. Two-tailed test determines if there is a relationship between variables in either direction.

When an alternative hypothesis has ‘≠’ sign, then a two-tailed test is performed. In contrast, when an alternative hypothesis has ‘> or <‘ sign, then one-tailed test is carried out.

**Alpha levels** sometimes called "Significance levels" It is probability of making the wrong decision when null hypothesis is true. One tail test has entire alpha level in one tail (either left or right). A two tailed test splits your alpha level in half.

### Errors in Hypothesis Testing

Whenever we reject Null hypothesis, there is a chance that we have made a mistake - i.e., that is we have rejected a true statement.

Rejecting a true null hypothesis is referred as **"Type I Error"**.

The probability of making such error is referred as **"Significance level"**.
It is denoted by **alpha (α)**

Failing to reject a false null hypotheis "**Type II Error**".

Probability of making Type II Error is denoted by **beta (β)**.

If we accepted true null hypothesis and rejected false null hypothesis, we acted correctly.

The probability of rejecting a false null hypothesis is "**Power of the Test**"

#### Trade-off

In Hypothesis testing there is tradeoff between Type I and Type II errors. For a given sample size, reducing the probability of Type I Error increases probability of Type II Error and vice-versa.

#### How statistical tests are performed

Before performing the test we need to formulate Null and alternate hypothesis.

A test is performed to reject a Null Hypothesis.

Perform the test and you will get a test statistic. 

Then calculate the critical values from the alpha levels.

Next, Compare the critical values with the test statistic.

If the test statistic is greater than critical value, Null hypotheis is rejected. If the test statistic is less than critical value we accept null hypothesis.

#### What is critical value

Critical Values are used in Hypothesis testing.

Critical value is the line that separates accepted region from rejected region.



Critical Value is calculated from alpha levels. For a confidence level of 95% alpha is 5%. The boundary which separates the accepted region of 95% and rejected region (Critical Region) of 5% is critical value.

#### P-value

P value is probability for the null hypothesis to be true. p-value reflects strength of null hypothesis.

#### t-test vs z-test

Z-test is used when "Population mean" and "Population Standard deviation" are known and sample size > 30.

Z-test follows Normal distribution.



T-test is used when we dont know population mean and population standard deviation and have to be estimated from the sample data. 

We use t-test when sample size < 30.

t-test follows student's T-distribution.




This is where the t-test comes into play. It helps us understand if the difference between two sample means is actually real or simply due to chance.

There are three different versions of t-test

1. Independent samples t-test

2. Paired sample t-test

3. One sample t-test

#### Independent sample t-test

*Compares mean for two groups.*

Example: Let’s say you’re curious about whether New Yorkers and Kansans spend a different amount of money per month on movies. It’s impractical to ask every New Yorker and Kansan about their movie spending, so instead you ask a sample of each—maybe 300 New Yorkers and 300 Kansans—and the means are $14 and $18. 

The t-test asks whether that difference is probably representative of a real difference between Kansans and New Yorkers generally or whether that is most likely a meaningless statistical fluk

#### Paried sample t-test

*Compares means from a same group at different times (say one year apart)*

Ex: two blood pressure measurements of the same person at at different hospitals.

#### One-sample t-test

*compares mean to the known standard*

comparing the mean volume of 10 water bottles to what is written on the label.

### ANOVA

Compares the means between the groups and it helps you to figure out if you need to reject null hypothesis or accept alternate hypothesis.



#### When to use ANOVA?

EX: A group of psychiatric patients are trying three different therapies. You need to see if one therapy is better than the other.

EX: A manufacturer has three different processes to make light bulbs. You need to know if one processes if better than the other.

#### Types of ANOVA

1. One-way ANOVA

2. Two-way ANOVA

3. Factorial ANOVA

4. MANOVA

**One-Way ANOVA:** There is only one Independent variable.

Ex: Test job satisfaction levels (IV) between ethnicities(DV).

**Two-way ANOVA:** You can have two of more independent variables. Two way ANOVA adds one more categorical independent variable


**Factorial ANOVA** Factorial ANOVA adds any number of categorical IVs to the regression.

**MANOVA:** Used when you have multiple dependent and independent variables.



They both are parametric statistical techniques. They both assume population from which sample is drawn is normally distributed.

 Chi square is a non-parametric test that is used to show association between two qualitative variables (like gender and eye color) ; while correlation (Pearson coefficient) is used to test the correlation between two quantitative variables (like heart rate and blood pressure)
