## Point and Interval Estimating

The estimation is used to make inferences about the population. There are two types of estimates, Point estimate and interval estimate.

A "Statistic" is characteristic of sample used to estimate the population parameter. In point estimation we use a "Single value" as a sample statistic to infer population parameter.

Interval estimate is a range of numbers in which population parameter lies.



Point Estimates are never perfect; they always have an error associated with it. We represent that error as "**Margin of Error**".

The error is expressed as a  **confidence interval**.
